```
import yfinance as yf
```

```
mbk = yf.Ticker('mbk.bk')
interval = '1d'
start = '2020-9-1'
end = '2023-9-7'
df = mbk.history(interval=interval,start=start,end=end)
df.head()
```

```
df = df.dropna()
df = df.iloc[:,:4]
df.head()
```

```
test = df['Close'].shift(-1)
test.head()
```

```
df['S_10'] = df['Close'].rolling(window=10).mean()
df['Corr'] = df['Close'].rolling(window=10).corr(df['S_10'])
df['Open-Close'] = df['Open'] - df['Close'].shift(1) #ราคาเปิดของวันปัจจุบัน ลบราคาปิดเมื่อวาน
df['Open-Open'] = df['Open'] - df['Open'].shift(1)
df = df.dropna()
X = df.iloc[:,:8]
X
```
